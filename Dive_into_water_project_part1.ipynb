```python
files_iter =[pd.read_csv("./MinneMUDAC_raw_files/"+ str(file) +"_metro_tax_parcels.txt", sep= "|", chunksize= 500) 
             for file in range(2002, 2016) if file != 2003]
first_chunks = [first(file).columns for file in files_iter]      
first_chunks
header_sets =[set(file) for file in first_chunks]
header_sets
             
```

```python
[(year, len(h)) for year, h in zip(range(2002, 2016), header_sets) if year != 2003]
```

```python
all_cols = set([])
for header_set in header_sets:
    all_cols = all_cols.union(header_set)
all_cols    
```

```python
common_cols = all_cols
for header_set in header_sets:
    common_cols = common_cols.intersection(header_set)
common_cols    
```

```python
len(common_cols)
```

```python
years = [year for year in range(2002, 2016) if year != 2003]
diff_cols =[header_set.difference(common_cols) for header_set in header_sets]   
extra_cols = [(file, len(set_cols)) for file, set_cols in zip(years,diff_cols)]
extra_cols
```
